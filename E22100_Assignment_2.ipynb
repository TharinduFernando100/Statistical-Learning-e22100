```python
from IPython.display import display, Markdown
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import beta

# Display formatted mathematical solution
solution = r"""
# PERT Distribution Model

The given parameters are

\[
a=8,\qquad b=20,\qquad c=11
\]

---

# Shape Parameters

The PERT shape parameters are

\[
\alpha = 1 + 4\left(\frac{c-a}{b-a}\right)
\]

Substituting values,

\[
\alpha = 1 + 4\left(\frac{11-8}{20-8}\right)
\]

\[
=1+4\left(\frac{3}{12}\right)
\]

\[
=1+1
\]

\[
\alpha =2
\]

Similarly,

\[
\beta = 1 + 4\left(\frac{b-c}{b-a}\right)
\]

\[
=1+4\left(\frac{20-11}{20-8}\right)
\]

\[
=1+4\left(\frac{9}{12}\right)
\]

\[
=1+3
\]

\[
\beta =4
\]

---

# Support of the Random Variable

Since \(X\) represents the task duration in days,

\[
8 \le X \le 20
\]

---

# Probability Density Function

The density function of a PERT distribution is

\[
f_X(x)=
\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}
{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}
\]

for \(a \le x \le b\).

Substituting \(\alpha=2\) and \(\beta=4\),

\[
f_X(x)=
\frac{(x-8)(20-x)^3}
{B(2,4)\times 12^5}
\]

---

# Beta Function

Using

\[
B(2,4)=\frac{\Gamma(2)\Gamma(4)}{\Gamma(6)}
\]

\[
=\frac{1!\times3!}{5!}
\]

\[
=\frac{6}{120}
\]

\[
=\frac{1}{20}
\]

Therefore,

\[
f_X(x)=
\frac{20(x-8)(20-x)^3}{12^5}
\]

for \(8 \le x \le 20\), and

\[
f_X(x)=0
\]

otherwise.

---

# Probability Space

A suitable probability space is

\[
\Omega=[8,20], \qquad \mathcal{F}=B([8,20])
\]

where \(B([8,20])\) is the Borel \(\sigma\)-algebra on \([8,20]\).

Define the probability measure by

\[
P(A)=\int_A f_X(x)\,dx,
\qquad A\in\mathcal{F}
\]

Define the random variable

\[
X:\Omega\to\mathbb{R}
\]

by

\[
X(\omega)=\omega
\]

Thus \(X\) records the task duration in days.

---

# Expectation

The expectation is

\[
E[X]=a+\frac{(b-a)\alpha}{\alpha+\beta}
\]

Substituting values,

\[
E[X]=8+\frac{(20-8)(2)}{2+4}
\]

\[
=8+\frac{24}{6}
\]

\[
=8+4
\]

\[
E[X]=12
\]

This is equivalent to the famous PERT mean formula

\[
E[X]=\frac{a+4c+b}{6}
\]

Substituting values,

\[
E[X]=\frac{8+4(11)+20}{6}
\]

\[
=\frac{72}{6}
\]

\[
E[X]=12
\]

---

# Variance

The variance approximation is

\[
Var(X)=\frac{(E[X]-a)(b-E[X])}{7}
\]

Substituting values,

\[
Var(X)=\frac{(12-8)(20-12)}{7}
\]

\[
=\frac{4\times8}{7}
\]

\[
=\frac{32}{7}
\]

\[
Var(X)\approx4.57
\]

---

# Standard Deviation

\[
\sigma=\sqrt{Var(X)}
\]

\[
=\sqrt{\frac{32}{7}}
\]

\[
\sigma\approx2.14
\]

Using the common PERT approximation,

\[
\sigma\approx\frac{b-a}{6}
\]

\[
=\frac{20-8}{6}
\]

\[
=2
\]

which is close to the exact value.

---

# Comparison with Triangular Distribution

| Quantity | Triangular | PERT |
|---|---|---|
| Mean | 13 | 12 |
| Variance | \(49/6 \approx 8.17\) | \(32/7 \approx 4.57\) |
| Standard deviation | \(\approx2.86\) | \(\approx2.14\) |
| Main feature | Linear, sharp peak | Smooth, mode weighted more heavily |

The PERT model gives a smoother distribution and pulls the expected value closer to the most likely value.

---

# Interpretation

- \(a=8\) is the shortest plausible completion time.
- \(b=20\) is the longest plausible completion time.
- \(c=11\) is the most likely completion time.
- \(E[X]=12\) is the expected task duration.
- \(\sigma\approx2.14\) measures uncertainty in task duration.

In project planning, these values help managers estimate schedules, assess deadline risk, and determine whether contingency time is needed.
"""

display(Markdown(solution))


import numpy as np
import plotly.graph_objects as go
from scipy.stats import triang, beta

# Parameters
a = 4
b = 10
c_mode = 6

# --- Triangular Distribution Calculation ---
# Scipy parametrization for Triangular
c_scipy = (c_mode - a) / (b - a)
loc = a
scale = b - a

# --- PERT Distribution Calculation ---
# PERT is a Beta distribution with specific alpha and beta parameters
# alpha = 1 + 4 * (c - a) / (b - a)
# beta = 1 + 4 * (b - c) / (b - a)
alpha_pert = 1 + 4 * (c_mode - a) / (b - a)
beta_pert = 1 + 4 * (b - c_mode) / (b - a)

# Generate data points
x = np.linspace(0, b, 500)
y_triang = triang.pdf(x, c=c_scipy, loc=loc, scale=scale)
y_pert = beta.pdf(x, alpha_pert, beta_pert, loc=a, scale=b-a)

# Create the Plotly figure
fig = go.Figure()

# Add Triangular Distribution
fig.add_trace(go.Scatter(
    x=x,
    y=y_triang,
    mode='lines',
    name='Triangular Distribution',
    line=dict(color='royalblue', width=2, dash='dash')
))

# Add PERT Distribution
fig.add_trace(go.Scatter(
    x=x,
    y=y_pert,
    mode='lines',
    name='PERT Distribution',
    line=dict(color='firebrick', width=2)
))

# Layout updates
fig.update_layout(
    title=f'Comparison: Triangular vs PERT (a={a}, b={b}, c={c_mode})',
    xaxis_title='Time (hours)',
    yaxis_title='Probability Density',
    template='plotly_white'
)

# Display the plot
fig.show()
```
