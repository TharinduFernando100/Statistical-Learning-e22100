```python
from IPython.display import display, Markdown

solution = r"""
# Determining Constants \(a\) and \(b\)

To find \(a\) and \(b\), we use the defining property of conditional expectation:

\[
\int_A E[X\mid Y]\,dP=\int_A X\,dP
\quad \text{for all } A \in \sigma(Y)
\]

---

# On the set \(\{\omega_1,\omega_2\}\)

\[
\int_{\{\omega_1,\omega_2\}} E[X\mid Y]\,dP
=
\int_{\{\omega_1,\omega_2\}} X\,dP
\]

### Left-hand side:

\[
a \cdot P(\{\omega_1,\omega_2\})
=
a(0.50+0.10)
=
0.60a
\]

### Right-hand side:

\[
0\cdot0.50 + 1\cdot0.10 = 0.10
\]

Thus,

\[
0.60a=0.10
\]

\[
a=\frac{0.10}{0.60}=\frac{1}{6}
\]

---

# On the set \(\{\omega_3,\omega_4\}\)

\[
\int_{\{\omega_3,\omega_4\}} E[X\mid Y]\,dP
=
\int_{\{\omega_3,\omega_4\}} X\,dP
\]

### Left-hand side:

\[
b \cdot P(\{\omega_3,\omega_4\})
=
b(0.20+0.20)
=
0.40b
\]

### Right-hand side:

\[
0\cdot0.20 + 1\cdot0.20 = 0.20
\]

Thus,

\[
0.40b=0.20
\]

\[
b=\frac{0.20}{0.40}=\frac{1}{2}
\]

---

# Final Result

\[
E[X\mid Y](\omega)=
\begin{cases}
\dfrac{1}{6}, & \omega\in\{\omega_1,\omega_2\},\\[6pt]
\dfrac{1}{2}, & \omega\in\{\omega_3,\omega_4\}
\end{cases}
\]

---

# As a Function of \(Y\)

\[
E[X\mid Y]
=
\frac{1}{6}\,\mathbf{1}_{\{Y=0\}}
+
\frac{1}{2}\,\mathbf{1}_{\{Y=1\}}
\]

or equivalently,

\[
E[X\mid Y]=
\begin{cases}
\dfrac{1}{6}, & Y=0,\\[6pt]
\dfrac{1}{2}, & Y=1
\end{cases}
\]

---

# Interpretation

- When \(Y=0\), the expected value of \(X\) is \(1/6\)
- When \(Y=1\), the expected value of \(X\) is \(1/2\)
- So \(E[X\mid Y]\) depends only on the value of \(Y\), as required
"""

display(Markdown(solution))
```

```python
from IPython.display import display, Markdown

solution = r"""
# Conditional Expectation from Conditional PMF

Using the discrete formula

\[
\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y),
\]

we compute the conditional expectation for each value of \(Y\).

---

# Evaluation at \(y=0\)

\[
\mathbb{E}[X\mid Y=0]
=
(0 \cdot p_{X|Y}(0 \mid 0))
+
(1 \cdot p_{X|Y}(1 \mid 0))
\]

\[
= 0 + (1 \cdot \tfrac{1}{6})
\]

\[
= \frac{1}{6}
\]

---

# Evaluation at \(y=1\)

\[
\mathbb{E}[X\mid Y=1]
=
(0 \cdot p_{X|Y}(0 \mid 1))
+
(1 \cdot p_{X|Y}(1 \mid 1))
\]

\[
= 0 + (1 \cdot \tfrac{1}{2})
\]

\[
= \frac{1}{2}
\]

---

# Final Result

\[
\mathbb{E}[X\mid Y]=
\begin{cases}
\dfrac{1}{6}, & Y=0,\\[6pt]
\dfrac{1}{2}, & Y=1
\end{cases}
\]

---

# Key Insight

This confirms the previous result:

- Conditional expectation is simply the **weighted average using conditional probabilities**.
- Since \(X \in \{0,1\}\), the expectation equals the probability that \(X=1\).

Thus,

\[
\mathbb{E}[X\mid Y=y] = P(X=1 \mid Y=y).
\]
"""

display(Markdown(solution))
```

```python
from IPython.display import display, Markdown

solution = r"""
# Minimum Mean Square Error (MMSE)

We compute the Minimum Mean Square Error (MMSE), which represents the optimal error when using the estimator \(\mathbb{E}[X \mid Y]\).

\[
\text{MSE}_{\min}
=
\mathbb{E}\big[(X-\mathbb{E}[X\mid Y])^2\big]
=
\mathbb{E}[\mathrm{Var}(X \mid Y)]
\]

---

# Step 1: Conditional Variance

Since \(X\) is Bernoulli given \(Y=y\), the variance is

\[
\mathrm{Var}(X \mid Y=y)=p(1-p)
\]

## For \(Y=0\):

\[
\mathrm{Var}(X \mid Y=0)
=
\frac{1}{6}\left(1-\frac{1}{6}\right)
=
\frac{1}{6}\cdot\frac{5}{6}
=
\frac{5}{36}
\]

## For \(Y=1\):

\[
\mathrm{Var}(X \mid Y=1)
=
\frac{1}{2}\left(1-\frac{1}{2}\right)
=
\frac{1}{2}\cdot\frac{1}{2}
=
\frac{1}{4}
\]

---

# Step 2: Take Expectation over \(Y\)

\[
\mathbb{E}[\mathrm{Var}(X \mid Y)]
=
\mathrm{Var}(X \mid Y=0)P(Y=0)
+
\mathrm{Var}(X \mid Y=1)P(Y=1)
\]

\[
=
\left(\frac{5}{36}\cdot0.60\right)
+
\left(\frac{1}{4}\cdot0.40\right)
\]

Convert probabilities:

\[
=
\left(\frac{5}{36}\cdot\frac{3}{5}\right)
+
\left(\frac{1}{4}\cdot\frac{2}{5}\right)
\]

\[
=
\frac{1}{12}
+
\frac{1}{10}
\]

\[
=
\frac{5+6}{60}
=
\frac{11}{60}
\]

---

# Final Result

\[
\text{MSE}_{\min}
=
\frac{11}{60}
\approx 0.1833
\]

---

# Interpretation

- This is the **lowest possible average squared error** achievable using \(Y\).
- It represents the **irreducible uncertainty** after using all information from the risk class.
- Any other estimator based on \(Y\) will have a strictly higher error.

---

# Key Insight

\[
\text{MMSE} = \mathbb{E}[\mathrm{Var}(X \mid Y)]
\]

This shows that the remaining error comes purely from **uncertainty within each risk group**, not from lack of information about \(Y\).
"""

display(Markdown(solution))
```

```python
from IPython.display import display, Markdown

solution = r"""
# Global Mean vs Conditional Estimation

If the bank ignores the risk class \(Y\) and uses the global mean

\[
\mathbb{E}[X]=0.30
\]

as its estimate, the resulting error is the **total variance of \(X\)**.

---

# Total Mean Square Error

When the estimate is a constant \(c=\mathbb{E}[X]\),

\[
\text{MSE}_{\text{global}}
=
\mathbb{E}[(X-\mathbb{E}[X])^2]
=
\mathrm{Var}(X)
\]

Since \(X\) is Bernoulli with \(p=0.30\),

\[
\mathrm{Var}(X)=p(1-p)=0.30(0.70)=0.21
\]

---

# Comparing the Errors

- **Optimal Error (with \(Y\))**:
\[
\text{MSE}_{\min}
=
\mathbb{E}[\mathrm{Var}(X\mid Y)]
=
\frac{11}{60}
\approx 0.1833
\]

- **Global Error (without \(Y\))**:
\[
\mathrm{Var}(X)=0.21
\]

---

# Improvement Using \(Y\)

Using the information in \(Y\), the bank reduces its prediction error:

\[
0.21 \;\rightarrow\; 0.1833
\]

---

# Law of Total Variance

The total uncertainty can be decomposed as

\[
\mathrm{Var}(X)
=
\mathbb{E}[\mathrm{Var}(X\mid Y)]
+
\mathrm{Var}(\mathbb{E}[X\mid Y])
\]

---

# Explained Variance

We compute

\[
\mathrm{Var}(\mathbb{E}[X\mid Y])
=
\mathbb{E}[(\mathbb{E}[X\mid Y])^2]
-
(\mathbb{E}[X])^2
\]

Since

\[
\mathbb{E}[X\mid Y]=
\begin{cases}
\frac{1}{6}, & \text{with probability } 0.6 \\
\frac{1}{2}, & \text{with probability } 0.4
\end{cases}
\]

we get

\[
\mathbb{E}[(\mathbb{E}[X\mid Y])^2]
=
\left(\frac{1}{36}\cdot 0.6\right)
+
\left(\frac{1}{4}\cdot 0.4\right)
=
0.0167 + 0.10
=
0.1167
\]

\[
(\mathbb{E}[X])^2 = (0.30)^2 = 0.09
\]

Thus,

\[
\mathrm{Var}(\mathbb{E}[X\mid Y])
=
0.1167 - 0.09
=
0.0267
\]

---

# Final Decomposition

\[
0.21
=
0.1833
+
0.0267
\]

---

# Interpretation

- **\(0.21\)** → Total uncertainty (no information used)
- **\(0.1833\)** → Remaining uncertainty after using \(Y\)
- **\(0.0267\)** → Reduction in error due to \(Y\)

---

# Key Insight

- The difference between global error and optimal error is exactly

\[
\mathrm{Var}(\mathbb{E}[X\mid Y])
\]

- This measures how much **predictive power** the variable \(Y\) provides.

- The better \(Y\) separates outcomes, the larger this term becomes.

---

# Summary

The Law of Total Variance shows:

- Using \(Y\) always reduces prediction error
- The remaining error is unavoidable noise
- The reduction quantifies the **value of information**
"""

display(Markdown(solution))
```

```python
from IPython.display import display, Markdown

solution = r"""
# Information-Theoretic Analysis

To calculate entropy and mutual information, we first summarize the joint and marginal distributions.

---

# 1. Joint Distribution \(P(X,Y)\)

\[
P(X=0,Y=0)=0.50,\quad
P(X=1,Y=0)=0.10
\]

\[
P(X=0,Y=1)=0.20,\quad
P(X=1,Y=1)=0.20
\]

---

# Marginal Distributions

- For \(Y\):
\[
P(Y=0)=0.60,\quad P(Y=1)=0.40
\]

- For \(X\):
\[
P(X=0)=0.70,\quad P(X=1)=0.30
\]

---

# 2. Conditional Entropy \(H(X \mid Y)\)

## Step A: Entropy given \(Y=0\)

\[
P(X=1 \mid Y=0)=\frac{1}{6},\quad
P(X=0 \mid Y=0)=\frac{5}{6}
\]

\[
H(X \mid Y=0)
=
-\left[
\frac{1}{6}\log_2\frac{1}{6}
+
\frac{5}{6}\log_2\frac{5}{6}
\right]
\approx 0.650 \text{ bits}
\]

---

## Step B: Entropy given \(Y=1\)

\[
P(X=1 \mid Y=1)=\frac{1}{2},\quad
P(X=0 \mid Y=1)=\frac{1}{2}
\]

\[
H(X \mid Y=1)
=
-\left[
\frac{1}{2}\log_2\frac{1}{2}
+
\frac{1}{2}\log_2\frac{1}{2}
\right]
=1.00 \text{ bit}
\]

---

## Step C: Total Conditional Entropy

\[
H(X \mid Y)
=
P(Y=0)H(X \mid Y=0)
+
P(Y=1)H(X \mid Y=1)
\]

\[
= (0.60 \times 0.650) + (0.40 \times 1.00)
\]

\[
= 0.390 + 0.40
= 0.790 \text{ bits}
\]

---

# 3. Mutual Information \(I(X;Y)\)

\[
I(X;Y)=H(X)-H(X \mid Y)
\]

## Step A: Entropy of \(X\)

\[
H(X)
=
-(0.70\log_2 0.70 + 0.30\log_2 0.30)
\approx 0.881 \text{ bits}
\]

---

## Step B: Final Mutual Information

\[
I(X;Y)
=
0.881 - 0.790
=
0.091 \text{ bits}
\]

---

# 4. Interpretation

- \(H(X)=0.881\): Total uncertainty about fraud before using \(Y\)
- \(H(X \mid Y)=0.790\): Remaining uncertainty after knowing risk class
- \(I(X;Y)=0.091\): Information gained from \(Y\)

---

# Key Insight

\[
\text{Information Gain} = \frac{0.091}{0.881} \approx 10.3\%
\]

- Knowing \(Y\) reduces uncertainty by about **10.3%**
- Since \(I(X;Y)>0\), the variable \(Y\) is informative

---

# Practical Meaning

Knowing whether a customer is "Low-Risk" or "High-Risk":

- does not eliminate uncertainty
- but provides **useful predictive signal**

Thus, \(Y\) is a valuable feature for fraud detection models.
"""

display(Markdown(solution))
```

```python
from IPython.display import display, Markdown

solution = r"""
# Joint Multivariate Normal Distribution

Let

\[
\mathbf{Z} =
\begin{bmatrix}
\mathbf{X} \\
\mathbf{Y}
\end{bmatrix}
\in \mathbb{R}^{n+m}
\]

---

# Definition

The random variable \(\mathbf{Z}\) is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

---

# Condition for a Joint Density

For \(\mathbf{Z}\) to possess a **joint density function** (i.e., be absolutely continuous with respect to Lebesgue measure on \(\mathbb{R}^{n+m}\)), its covariance matrix must be non-singular (positive definite).

---

# Joint Density Function

The joint density function of \((\mathbf{X}, \mathbf{Y})\) is

\[
f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) =
\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}}
\exp\left(
-\frac{1}{2}
(\mathbf{z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{z}-\boldsymbol{\mu})
\right)
\]

---

# Mean Vector

\[
\boldsymbol{\mu} = \mathbb{E}[\mathbf{Z}] =
\begin{bmatrix}
\boldsymbol{\mu}_X \\
\boldsymbol{\mu}_Y
\end{bmatrix}
\in \mathbb{R}^{n+m}
\]

where:
- \(\boldsymbol{\mu}_X \in \mathbb{R}^n\)
- \(\boldsymbol{\mu}_Y \in \mathbb{R}^m\)

---

# Covariance Matrix

\[
\boldsymbol{\Sigma} =
\begin{bmatrix}
\boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\
\boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY}
\end{bmatrix}
\]

where:

- \(\boldsymbol{\Sigma}_{XX} = \mathrm{Var}(\mathbf{X}) \in \mathbb{R}^{n \times n}\)
- \(\boldsymbol{\Sigma}_{YY} = \mathrm{Var}(\mathbf{Y}) \in \mathbb{R}^{m \times m}\)
- \(\boldsymbol{\Sigma}_{XY} = \mathrm{Cov}(\mathbf{X},\mathbf{Y})\)
- \(\boldsymbol{\Sigma}_{YX} = \boldsymbol{\Sigma}_{XY}^T\)

---

# Normalization Factor

\[
|\boldsymbol{\Sigma}|
\]

is the determinant of the covariance matrix and ensures that the density integrates to 1.

---

# Key Insight

- If \(\boldsymbol{\Sigma}\) is **positive definite**, the distribution is well-defined with a proper density.
- The block structure captures both:
  - internal variability (\(\Sigma_{XX}, \Sigma_{YY}\))
  - cross-dependence (\(\Sigma_{XY}\))
"""

display(Markdown(solution))
```

Marginal Distributions

```python
from IPython.display import display, Markdown

solution = r"""
# Marginal Density from Joint Gaussian Distribution

To derive the marginal density of \(\mathbf{X}\) (or \(\mathbf{Y}\)) from the joint density of \((\mathbf{X}, \mathbf{Y})\), we integrate over the other variable.

---

# Joint Density Structure

The joint density is

\[
f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})
\propto
\exp\left(-\frac{1}{2}Q\right),
\]

where the quadratic form is

\[
Q =
\begin{bmatrix}
\mathbf{x}-\boldsymbol{\mu}_x \\
\mathbf{y}-\boldsymbol{\mu}_y
\end{bmatrix}^T
\boldsymbol{\Sigma}^{-1}
\begin{bmatrix}
\mathbf{x}-\boldsymbol{\mu}_x \\
\mathbf{y}-\boldsymbol{\mu}_y
\end{bmatrix}.
\]

---

# Block Covariance Matrix

\[
\boldsymbol{\Sigma} =
\begin{bmatrix}
\boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\
\boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY}
\end{bmatrix}
\]

---

# Schur Complement (Key Step)

Using block matrix inversion, we define the Schur complement:

\[
\mathbf{S}_{YY}
=
\boldsymbol{\Sigma}_{YY}
-
\boldsymbol{\Sigma}_{YX}
\boldsymbol{\Sigma}_{XX}^{-1}
\boldsymbol{\Sigma}_{XY}
\]

---

# Inverse Structure of \(\boldsymbol{\Sigma}\)

The inverse can be expressed as

\[
\boldsymbol{\Sigma}^{-1}
=
\begin{bmatrix}
\mathbf{I} & -\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY} \\
\mathbf{0} & \mathbf{I}
\end{bmatrix}^T
\begin{bmatrix}
\boldsymbol{\Sigma}_{XX}^{-1} & \mathbf{0} \\
\mathbf{0} & \mathbf{S}_{YY}^{-1}
\end{bmatrix}
\begin{bmatrix}
\mathbf{I} & -\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY} \\
\mathbf{0} & \mathbf{I}
\end{bmatrix}
\]

---

# Interpretation

This factorization is crucial because:

- It separates the dependence structure between \(\mathbf{X}\) and \(\mathbf{Y}\)
- It allows **completing the square** in the exponent
- It isolates a Gaussian form in \(\mathbf{X}\)

---

# Key Result Idea

After completing the square and integrating over \(\mathbf{Y}\):

\[
\mathbf{X} \sim \mathcal{N}(\boldsymbol{\mu}_X, \boldsymbol{\Sigma}_{XX})
\]

and similarly

\[
\mathbf{Y} \sim \mathcal{N}(\boldsymbol{\mu}_Y, \boldsymbol{\Sigma}_{YY})
\]

---

# Insight

- The marginal distribution of a multivariate normal is still **multivariate normal**
- Cross-covariance terms affect conditioning, not marginalization
"""

display(Markdown(solution))
```

Marginal Density of  X

```python
from IPython.display import display, Markdown

solution = r"""
# Marginal Density of \(\mathbf{X}\) from Joint Gaussian

Let

\[
\widetilde{\mathbf{x}} = \mathbf{x} - \boldsymbol{\mu}_x,
\quad
\widetilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_y
\]

---

# Quadratic Form Decomposition

The exponent can be decomposed as:

\[
Q =
\widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}}
+
(\widetilde{\mathbf{y}} - \mathbf{A}\widetilde{\mathbf{x}})^T
\mathbf{S}_{YY}^{-1}
(\widetilde{\mathbf{y}} - \mathbf{A}\widetilde{\mathbf{x}})
\]

where

\[
\mathbf{A} = \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}
\]

---

# Factorized Joint Density

\[
f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})
=
C \cdot
\exp\left(
-\frac{1}{2}
\widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}}
\right)
\cdot
\exp\left(
-\frac{1}{2}
(\widetilde{\mathbf{y}} - \mathbf{m})^T
\mathbf{S}_{YY}^{-1}
(\widetilde{\mathbf{y}} - \mathbf{m})
\right)
\]

where

\[
\mathbf{m} = \mathbf{A}\widetilde{\mathbf{x}}
\]

---

# Marginal Density Definition

\[
f_{\mathbf{X}}(\mathbf{x})
=
\int_{\mathbb{R}^m}
f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})
\, d\mathbf{y}
\]

Pulling out \(\mathbf{x}\)-dependent terms:

\[
f_{\mathbf{X}}(\mathbf{x})
=
C \cdot
\exp\left(
-\frac{1}{2}
\widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}}
\right)
\int_{\mathbb{R}^m}
\exp\left(
-\frac{1}{2}
(\widetilde{\mathbf{y}} - \mathbf{m})^T
\mathbf{S}_{YY}^{-1}
(\widetilde{\mathbf{y}} - \mathbf{m})
\right)
d\mathbf{y}
\]

---

# Gaussian Integral Result

The integral is the normalization constant of a multivariate Gaussian:

\[
\int_{\mathbb{R}^m} (\cdots) d\mathbf{y}
=
\sqrt{(2\pi)^m |\mathbf{S}_{YY}|}
\]

---

# Determinant Identity

\[
|\boldsymbol{\Sigma}|
=
|\boldsymbol{\Sigma}_{XX}|
\cdot
|\mathbf{S}_{YY}|
\]

---

# Final Simplification

Combining constants:

\[
f_{\mathbf{X}}(\mathbf{x})
=
\frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{XX}|}}
\exp\left(
-\frac{1}{2}
(\mathbf{x}-\boldsymbol{\mu}_x)^T
\boldsymbol{\Sigma}_{XX}^{-1}
(\mathbf{x}-\boldsymbol{\mu}_x)
\right)
\]

---

# Final Result

\[
\mathbf{X} \sim \mathcal{N}(\boldsymbol{\mu}_x, \boldsymbol{\Sigma}_{XX})
\]

---

# Key Conclusion

- Marginal of a joint Gaussian is Gaussian
- Cross-covariance affects only conditioning, not marginalization
- Block matrix structure ensures consistency of Gaussian family
"""

display(Markdown(solution))
```

Marginal Density of  Y

```python
from IPython.display import display, Markdown

solution = r"""
# Marginal Density of \(\mathbf{Y}\) from Joint Gaussian

The derivation for \(f_{\mathbf{Y}}(\mathbf{y})\) is symmetric to that of \(\mathbf{X}\). We integrate out \(\mathbf{x}\) from the joint density.

---

# Quadratic Form Decomposition

Using the Schur complement of \(\boldsymbol{\Sigma}_{YY}\), the quadratic form becomes:

\[
Q =
\widetilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \widetilde{\mathbf{y}}
+
(\widetilde{\mathbf{x}} - \mathbf{B}\widetilde{\mathbf{y}})^T
\mathbf{S}_{XX}^{-1}
(\widetilde{\mathbf{x}} - \mathbf{B}\widetilde{\mathbf{y}})
\]

where

\[
\widetilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_y
\]

\[
\mathbf{B} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
\]

\[
\mathbf{S}_{XX}
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\]

---

# Marginal Density Definition

\[
f_{\mathbf{Y}}(\mathbf{y})
=
\int_{\mathbb{R}^n}
f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})
\, d\mathbf{x}
\]

---

# Separation of Terms

The term

\[
\exp\left(
-\frac{1}{2}
\widetilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \widetilde{\mathbf{y}}
\right)
\]

is independent of \(\mathbf{x}\), so it factors out:

\[
f_{\mathbf{Y}}(\mathbf{y})
=
C \cdot
\exp\left(
-\frac{1}{2}
\widetilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \widetilde{\mathbf{y}}
\right)
\int_{\mathbb{R}^n}
\exp\left(
-\frac{1}{2}
(\widetilde{\mathbf{x}} - \mathbf{B}\widetilde{\mathbf{y}})^T
\mathbf{S}_{XX}^{-1}
(\widetilde{\mathbf{x}} - \mathbf{B}\widetilde{\mathbf{y}})
\right)
d\mathbf{x}
\]

---

# Gaussian Integral Result

The integral is the normalization constant of an \(n\)-dimensional Gaussian:

\[
\int_{\mathbb{R}^n} (\cdots)\, d\mathbf{x}
=
\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}
\]

---

# Determinant Identity

\[
|\boldsymbol{\Sigma}|
=
|\boldsymbol{\Sigma}_{YY}|
\cdot
|\mathbf{S}_{XX}|
\]

---

# Final Simplification

All constants cancel, yielding:

\[
f_{\mathbf{Y}}(\mathbf{y})
=
\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}
\exp\left(
-\frac{1}{2}
(\mathbf{y}-\boldsymbol{\mu}_y)^T
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y)
\right)
\]

---

# Final Result

\[
\mathbf{Y} \sim \mathcal{N}(\boldsymbol{\mu}_y, \boldsymbol{\Sigma}_{YY})
\]

---

# Key Insight

- Marginals of a joint Gaussian remain Gaussian
- Symmetry between \(\mathbf{X}\) and \(\mathbf{Y}\)
- Schur complements determine conditional structure, not marginal form
"""

display(Markdown(solution))
```

Conditional Density

```python
from IPython.display import display, Markdown

solution = r"""
# Conditional Density of \(\mathbf{X} \mid \mathbf{Y}\)

To derive the conditional density \(f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})\), we use:

\[
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})
=
\frac{f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}
\]

---

# Quadratic Form Decomposition

From previous results:

\[
Q_{joint}
=
(\mathbf{y}-\boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y}-\boldsymbol{\mu}_y)
+
(\mathbf{x}-\boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x}-\boldsymbol{\mu}^*)
\]

where

\[
\mathbf{S}_{XX}
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\]

\[
\boldsymbol{\mu}^*
=
\boldsymbol{\mu}_x
+
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y)
\]

---

# Marginal Exponent

\[
Q_{marginal}
=
(\mathbf{y}-\boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y}-\boldsymbol{\mu}_y)
\]

---

# Exponent Subtraction

\[
Q_{conditional}
=
Q_{joint} - Q_{marginal}
=
(\mathbf{x}-\boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x}-\boldsymbol{\mu}^*)
\]

---

# Constant Terms

Joint constant:

\[
\frac{1}{\sqrt{(2\pi)^{n+m}|\boldsymbol{\Sigma}|}}
\]

Marginal constant:

\[
\frac{1}{\sqrt{(2\pi)^m|\boldsymbol{\Sigma}_{YY}|}}
\]

Ratio:

\[
\frac{1}{\sqrt{(2\pi)^n \frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}}}
\]

Using:

\[
|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|
\]

we get:

\[
\frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|} = |\mathbf{S}_{XX}|
\]

---

# Final Conditional Density

\[
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})
=
\frac{1}{\sqrt{(2\pi)^n|\mathbf{S}_{XX}|}}
\exp\left(
-\frac{1}{2}
(\mathbf{x}-\boldsymbol{\mu}^*)^T
\mathbf{S}_{XX}^{-1}
(\mathbf{x}-\boldsymbol{\mu}^*)
\right)
\]

---

# Final Result

\[
(\mathbf{X} \mid \mathbf{Y}=\mathbf{y})
\sim
\mathcal{N}
\left(
\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y}-\boldsymbol{\mu}_y),
\;
\boldsymbol{\Sigma}_{XX}-\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\right)
\]

---

# Summary

- Conditional mean shifts with observation \(y\)
- Conditional covariance is independent of \(y\)
- Gaussian distributions remain closed under conditioning
"""

display(Markdown(solution))
```

Conditional Expectation

```python
from IPython.display import display, Markdown

solution = r"""
# Conditional Expectation in the Multivariate Normal Case

Let
\[
g(y) = \mathbb{E}[X \mid Y=y]
\]

By the Disintegration Theorem,
\[
g(y)
=
\int_{\mathbb{R}^n}
\mathbf{x}\,
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})
\, d\mathbf{x}
\]

---

# Interpretation

This is the first moment (mean) of the conditional Gaussian distribution:

\[
(\mathbf{X} \mid \mathbf{Y}=\mathbf{y})
\sim
\mathcal{N}\left(
\boldsymbol{\mu}_x
+
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y),
\;
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\right)
\]

---

# Therefore

The conditional expectation is simply the conditional mean:

\[
g(y)
=
\boldsymbol{\mu}_x
+
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y)
\]

---

# Final Result

\[
\mathbb{E}[X \mid Y=y]
=
\boldsymbol{\mu}_x
+
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y)
\]

---

# Random Variable Form

Replacing \(y\) with \(Y\):

\[
\mathbb{E}[X \mid Y]
=
\boldsymbol{\mu}_x
+
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{Y}-\boldsymbol{\mu}_y)
\]

---

# Key Insight

- Conditional expectation is linear in \(Y\)
- It is the mean of the conditional Gaussian distribution
- Acts as the optimal MMSE estimator in Gaussian systems
"""

display(Markdown(solution))
```

The Conditional Variances

```python
from IPython.display import display, Markdown

solution = r"""
# Conditional Variance and Total Variance Decomposition (Gaussian Case)

---

# Conditional Distribution

Since
\[
(\mathbf{X} \mid \mathbf{Y}=\mathbf{y})
\sim
\mathcal{N}\left(
\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y}-\boldsymbol{\mu}_y),
\;
\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\right)
\]

---

# Conditional Variance

\[
\mathrm{Var}(\mathbf{X} \mid \mathbf{Y}=\mathbf{y})
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\]

Since this does not depend on \(y\),

\[
\mathrm{Var}(\mathbf{X} \mid \mathbf{Y})
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\]

---

# Expected Conditional Variance (MMSE Error)

\[
\mathbb{E}[\mathrm{Var}(\mathbf{X} \mid \mathbf{Y})]
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\]

Thus the minimum mean square estimation error is:

\[
\mathbb{E}\left[(\mathbf{X}-\mathbb{E}[\mathbf{X}|\mathbf{Y}])^2\right]
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\]

---

# Total Variance Decomposition

We use the law of total variance:

\[
\mathrm{Var}(\mathbf{X})
=
\mathbb{E}[\mathrm{Var}(\mathbf{X}|\mathbf{Y})]
+
\mathrm{Var}(\mathbb{E}[\mathbf{X}|\mathbf{Y}])
\]

---

# Substituting Gaussian Results

\[
\mathrm{Var}(\mathbf{X})
=
\left(
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\right)
+
\mathrm{Var}(\mathbb{E}[\mathbf{X}|\mathbf{Y}])
\]

Rearranging:

\[
\mathrm{Var}(\mathbb{E}[\mathbf{X}|\mathbf{Y}])
=
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\]

---

# Final Interpretation

## Explained Variance
\[
\mathrm{Var}(\mathbb{E}[\mathbf{X}|\mathbf{Y}])
=
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\]

Represents:
- Portion of uncertainty in \(X\)
- Explained by linear relationship with \(Y\)

---

## Unexplained Variance (Residual Error)

\[
\mathrm{Var}(\mathbf{X}|\mathbf{Y})
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\]

Represents:
- Remaining uncertainty after observing \(Y\)
- Irreducible estimation error

---

# Key Insight

The total uncertainty in \(X\) splits into:

- **Explained part** (from \(Y\))
- **Unexplained part** (noise / residual)
"""

display(Markdown(solution))
```

The Differential Entropy of a Multivaraite Gaussian

```python
from IPython.display import display, Markdown

solution = r"""
# Differential Entropy of a Multivariate Normal Distribution

Let
\[
\mathbf{Z} =
\begin{bmatrix}
\mathbf{X} \\
\mathbf{Y}
\end{bmatrix}
\in \mathbb{R}^{n+m}
\]

with mean \(\boldsymbol{\mu}\) and covariance \(\boldsymbol{\Sigma}\).

The joint density is:

\[
f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})
=
\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}}
\exp\left(
-\frac{1}{2}
(\mathbf{z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{z}-\boldsymbol{\mu})
\right)
\]

---

# Differential Entropy Definition

\[
h(\mathbf{Z}) = -\mathbb{E}[\ln f_{\mathbf{Z}}(\mathbf{z})]
\]

---

# Step 1: Log Density

\[
\ln f_{\mathbf{Z}}(\mathbf{z})
=
-\frac{1}{2}\ln\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right)
-\frac{1}{2}
(\mathbf{z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{z}-\boldsymbol{\mu})
\]

---

# Step 2: Apply Expectation

\[
h(\mathbf{Z})
=
\frac{1}{2}\ln\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right)
+
\frac{1}{2}
\mathbb{E}
\left[
(\mathbf{Z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{Z}-\boldsymbol{\mu})
\right]
\]

---

# Step 3: Quadratic Form Expectation

For a Gaussian random vector:

\[
\mathbb{E}
\left[
(\mathbf{Z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{Z}-\boldsymbol{\mu})
\right]
=
\mathrm{Tr}(\mathbf{I}_{n+m})
=
n+m
\]

---

# Step 4: Substitute Back

\[
h(\mathbf{Z})
=
\frac{1}{2}\ln\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right)
+
\frac{n+m}{2}
\]

---

# Final Step

\[
h(\mathbf{Z})
=
\frac{1}{2}
\ln\left((2\pi e)^{n+m}|\boldsymbol{\Sigma}|\right)
\]

---

# Final Result

\[
\boxed{
h(\mathbf{Z})
=
\frac{1}{2}
\ln\left((2\pi e)^{n+m}|\boldsymbol{\Sigma}|\right)
}
\]

---

# Key Insight

- Entropy depends only on covariance \(\boldsymbol{\Sigma}\)
- Higher determinant ⇒ higher uncertainty
- Independent of mean \(\boldsymbol{\mu}\)
"""

display(Markdown(solution))
```

Conditional Entropy

```python
from IPython.display import display, Markdown

solution = r"""
# Conditional Entropy of a Multivariate Gaussian

The conditional entropy \(h(\mathbf{X} \mid \mathbf{Y})\) measures the remaining uncertainty in \(\mathbf{X}\) after observing \(\mathbf{Y}\).

---

# Conditional Distribution

For a joint Gaussian:

\[
(\mathbf{X} \mid \mathbf{Y}=\mathbf{y})
\sim
\mathcal{N}\left(
\boldsymbol{\mu}_x
+
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y}-\boldsymbol{\mu}_y),
\;
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\right)
\]

---

# Conditional Covariance

The conditional covariance is:

\[
\boldsymbol{\Sigma}_{X|Y}
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\]

---

# Key Property

\[
\boldsymbol{\Sigma}_{X|Y} \text{ is independent of } \mathbf{y}
\]

So the uncertainty is constant for all observations of \(Y\).

---

# Conditional Entropy Formula

For a multivariate Gaussian:

\[
h(\mathbf{X}|\mathbf{Y})
=
\frac{1}{2}
\ln\left(
(2\pi e)^n
\left|
\boldsymbol{\Sigma}_{X|Y}
\right|
\right)
\]

---

# Final Result

\[
h(\mathbf{X}|\mathbf{Y})
=
\frac{1}{2}
\ln\left(
(2\pi e)^n
\left|
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\right|
\right)
\]

---

# Key Insight

- Conditioning reduces entropy
- Reduction depends on correlation between \(X\) and \(Y\)
- If independent ⇒ no entropy reduction
"""

display(Markdown(solution))
```

The Entropy of the Conditional Expectation

```python
from IPython.display import display, Markdown

solution = r"""
# Differential Entropy of the Conditional Mean (Explained Component)

From the law of total variance:

\[
\mathrm{Var}(\mathbb{E}[\mathbf{X} \mid \mathbf{Y}])
=
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\]

---

# Interpretation

The random variable \(\mathbb{E}[\mathbf{X} \mid \mathbf{Y}]\) is a linear transformation of \(\mathbf{Y}\), and in the Gaussian case it is also Gaussian.

Thus:

\[
\mathbb{E}[\mathbf{X} \mid \mathbf{Y}]
\sim
\mathcal{N}
\left(
\boldsymbol{\mu}_x,
\;
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\right)
\]

---

# Differential Entropy Formula

For a multivariate Gaussian:

\[
h(\mathbf{Z})
=
\frac{1}{2}
\ln\left((2\pi e)^n |\boldsymbol{\Sigma}|\right)
\]

---

# Apply to the Explained Variance Component

\[
h(\mathbb{E}[\mathbf{X} \mid \mathbf{Y}])
=
\frac{1}{2}
\ln\left(
(2\pi e)^n
\left|
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\right|
\right)
\]

---

# Final Result

\[
\boxed{
h(\mathbb{E}[\mathbf{X} \mid \mathbf{Y}])
=
\frac{1}{2}
\ln\left(
(2\pi e)^n
\left|
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\right|
\right)
}
\]

---

# Key Insight

- This entropy measures the uncertainty in the *predictable part* of \(\mathbf{X}\)
- It depends only on correlation between \(X\) and \(Y\)
- If \(X\) and \(Y\) are independent ⇒ entropy becomes zero (no explained variation)
"""

display(Markdown(solution))
```

Mutual Information

```python
from IPython.display import display, Markdown

solution = r"""
# Mutual Information for Multivariate Gaussian Variables

---

# Definition

\[
I(\mathbf{X};\mathbf{Y})
=
h(\mathbf{X}) - h(\mathbf{X} \mid \mathbf{Y})
\]

---

# Step 1: Substitute Entropy Expressions

\[
h(\mathbf{X})
=
\frac{1}{2}
\ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{XX}|\right)
\]

\[
h(\mathbf{X} \mid \mathbf{Y})
=
\frac{1}{2}
\ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{X|Y}|\right)
\]

---

# Step 2: Compute Mutual Information

\[
I(\mathbf{X};\mathbf{Y})
=
\frac{1}{2}
\ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{XX}|\right)
-
\frac{1}{2}
\ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{X|Y}|\right)
\]

---

# Step 3: Cancel Common Terms

Using:

\[
\ln A - \ln B = \ln\left(\frac{A}{B}\right)
\]

we get:

\[
I(\mathbf{X};\mathbf{Y})
=
\frac{1}{2}
\ln\left(
\frac{|\boldsymbol{\Sigma}_{XX}|}{|\boldsymbol{\Sigma}_{X|Y}|}
\right)
\]

---

# Step 4: Substitute Conditional Covariance

\[
\boldsymbol{\Sigma}_{X|Y}
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\]

---

# Final Result

\[
\boxed{
I(\mathbf{X};\mathbf{Y})
=
\frac{1}{2}
\ln\left(
\frac{
|\boldsymbol{\Sigma}_{XX}|
}{
|\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}|
}
\right)
}
\]

---

# Key Insight

- Mutual information measures reduction in uncertainty
- Depends only on covariance structure
- If \(X\) and \(Y\) are independent ⇒ \(I(X;Y)=0\)
"""

display(Markdown(solution))
```

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import multivariate_normal, norm

# --- 1. Parameters for the Spindle Problem ---
# X: Temperature ~ N(0, 1)
# Q: Torque ~ N(0, 4)
# Y: Power = X + Q => Y ~ N(0, 5)
mu_x, mu_y = 0, 0
var_x, var_q = 1, 4
var_y = var_x + var_q
cov_xy = var_x  # Cov(X, X+Q) = Var(X) = 1
rho = cov_xy / (np.sqrt(var_x) * np.sqrt(var_y))

# Shrinkage Factor (Gain)
k = cov_xy / var_y  # 1/5 = 0.2

# --- 2. Generate Data for Visualization ---
# Sample points to show the "Cloud" of data
np.random.seed(42)
n_samples = 1000
x_samples = np.random.normal(mu_x, np.sqrt(var_x), n_samples)
q_samples = np.random.normal(0, np.sqrt(var_q), n_samples)
y_samples = x_samples + q_samples
x_hat = k * y_samples  # E[X|Y]
residual = x_samples - x_hat

# Grids for Contours
x_grid = np.linspace(-4, 4, 200)
y_grid = np.linspace(-8, 8, 200)
X_m, Y_m = np.meshgrid(x_grid, y_grid)
pos = np.dstack((X_m, Y_m))
Z = multivariate_normal([mu_x, mu_y], [[var_x, cov_xy], [cov_xy, var_y]]).pdf(pos)

# --- 3. Create Simulation Plot ---
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "Joint Distribution: X vs Y<br><sup>The Line of Best Estimate (Shrinkage)</sup>",
        "Estimation Error (Residual) vs Y<br><sup>Visualizing the Orthogonality Principle</sup>"
    )
)

# --- LEFT PLOT: Joint X, Y and the Estimator ---
# Density Contour
fig.add_trace(go.Contour(
    x=x_grid, y=y_grid, z=Z, colorscale='Blues', showscale=False, opacity=0.6,
    hoverinfo='skip'
), row=1, col=1)

# Scatter of actual realizations
fig.add_trace(go.Scatter(
    x=x_samples, y=y_samples, mode='markers',
    marker=dict(color='rgba(0,0,0,0.1)', size=3),
    name="Realized (x, y)"
), row=1, col=1)

# The Best Estimate Line E[X|Y] = 0.2Y
y_line = np.linspace(-8, 8, 10)
fig.add_trace(go.Scatter(
    x=k * y_line, y=y_line,
    line=dict(color='Red', width=4),
    name="E[X|Y] = 0.2Y (Optimal)"
), row=1, col=1)

# The Naive Estimator Line X = Y
fig.add_trace(go.Scatter(
    x=y_line, y=y_line,
    line=dict(color='Black', width=2, dash='dash'),
    name="X = Y (Naive)"
), row=1, col=1)

# --- RIGHT PLOT: Orthogonality (Residual vs Y) ---
fig.add_trace(go.Scatter(
    x=y_samples, y=residual, mode='markers',
    marker=dict(color='Purple', opacity=0.4, size=4),
    name="Residual (X - 0.2Y)"
), row=1, col=2)

# Horizontal line at 0 to show no correlation
fig.add_trace(go.Scatter(
    x=y_line, y=np.zeros_like(y_line),
    line=dict(color='Black', width=2),
    name="Zero Mean Error"
), row=1, col=2)

# --- Layout and Styling ---
fig.update_layout(
    height=600,
    title_text="Spindle Motor Diagnostic Simulation: Extracting Signal from Noise",
    template="plotly_white",
    hovermode='x'
)

fig.update_xaxes(title_text="True Temperature (X)", row=1, col=1)
fig.update_yaxes(title_text="Observed Power (Y)", row=1, col=1)
fig.update_xaxes(title_text="Observed Power (Y)", row=1, col=2)
fig.update_yaxes(title_text="Residual Error (Z)", row=1, col=2)

fig.show()
```
